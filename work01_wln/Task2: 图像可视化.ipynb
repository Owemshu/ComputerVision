```python
import matplotlib.image as image
import matplotlib.pyplot as plt
```

```python
# 1.利用matplotlib库可视化图像以及对应的RGB三通道
img = image.imread("./img/SnowGirl.jpg")
# 分离 RGB
img_R = img * [1, 0, 0]
img_G = img * [0, 1, 0]
img_B = img * [0, 0, 1]
# 输出
plt.subplot(2, 2, 1)
plt.imshow(img)
plt.axis('off')
plt.subplot(2, 2, 2)
plt.imshow(img_R)
plt.axis('off')
plt.subplot(2, 2, 3)
plt.imshow(img_G)
plt.axis('off')
plt.subplot(2, 2, 4)
plt.imshow(img_B)
plt.axis('off')
```
